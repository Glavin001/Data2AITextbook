```python
%%script false --no-raise-error

persona = """
You are an expert educator with mastery of Bloom's Taxonomy (types of knowledge and cognitive processes).
Your objective is to teach a student to know, understand, and/or use all knowledge contained in the following input context passage.
The student should be able to recall, recognize, and/or locate all knowledge contained in the input context passage.
"""

def extract_knowledge(input: str):
    # Step 1: send the conversation and available functions to GPT
    task = """Your first task is to produce a list of all knowledge contained in the input text.
Output all / every single topic/knowledge included in a provided input text grouped by the knowledge dimensions (factual, conceptual, procedural, metacognitive).
Do not miss any topics.
You do not need to include the entire sentence for each of the topics, use a short form as a reminder.
Later you'll be given the input text again with only 1 of the topics to elaborate on in detail.
Do not add new information not captured in the input text.
You can interpret, rephrase, etc however ensure all output is strongly grounded in the input text.
"""
    prompt = f"""{persona}
{task}

Input Text:
{input}
"""
    messages = [
        {"role": "user", "content": prompt}
    ]
    functions = [
        {
            "name": "write_knowledge",
            "description": "Write an exhaustive list of every piece of knowledge/topic/subject within the input text.",
            "parameters": {
                "type": "object",
                "properties": {
                    # "notes": {
                    #     "type": "string",
                    #     "description": "Write notes here. Take a deep breath and think step by step about what you need to do.",
                    # },
                    "learning_objectives": {
                        "type": "array",
                        "description": "All knowledge/topics/subjects within the input text. Should be unique, combine duplicates or very similar knowledge",
                        "items": {
                            "type": "object",
                            "properties": {
                                # "title": {
                                "description": {
                                    "type": "string",
                                    "description": "A short distinctive informative description of the learning objective/knowledge/topic/subject. Concise, remove unnecessary/stop words. Information dense/rich",
                                },
                                # "related_topics": {"type": "array", "items": {"type": "string"}},
                                # "knowledge_types": {"type": "array", "items": {
                                #     "type": "object",
                                #     "properties": {
                                #         "knowledge_type": {"type": "string"},
                                #         "cognitive_process": {"type": "string"},
                                #     },
                                # }},
                                # Enums
                                # "knowledge_types": {
                                #     "type": "array",
                                #     "minItems": 1,
                                #     "items": {
                                #         "type": "string",
                                #         "description": "All Bloom's Taxonomy Knowledge Types that are related to the knowledge",
                                #         "enum": knowledge_types,
                                #     }
                                # },
                                # Knowledge Types except it is an object with a property for each in knowledge_types
                                # Generate with a template by iterating over knowledge_types
                                "knowledge_types": {
                                    "type": "object",
                                    # iterate over knowledge_types and create a property for each
                                    # using reduce
                                    "description": "All Bloom's Taxonomy Knowledge Types this knowledge could be used for",
                                    "properties": {
                                        knowledge_type: {
                                            "type": "boolean",
                                            # "description": f"Whether this knowledge is a {knowledge_type} type of knowledge",
                                        } for knowledge_type in knowledge_types
                                    }
                                },
                                # "cognitive_processes": {
                                #     "type": "array",
                                #     "minItems": 1,
                                #     "items": {
                                #         "type": "string",
                                #         "description": "All Bloom's Taxonomy Cognitive Processes that are related to the knowledge",
                                #         "enum": cognitive_processes,
                                #     }
                                # },
                                # Cognitive Processes except it is an object with a property for each in cognitive_processes
                                # Generate with a template by iterating over cognitive_processes
                                "cognitive_processes": {
                                    "type": "object",
                                    # iterate over cognitive_processes and create a property for each
                                    # using reduce
                                    "description": "All Bloom's Taxonomy Cognitive Processes this knowledge could be used for",
                                    "properties": {
                                        cognitive_process: {
                                            "type": "boolean",
                                            # "description": f"Whether this knowledge is a {cognitive_process} type of knowledge",
                                        } for cognitive_process in cognitive_processes
                                    }
                                },
                                "verbs": {
                                    "type": "array",
                                    "description": "Exhaustive list of all Bloom's Taxonomy action verbs related to this knowledge which can be used to write lessons/exercises for the learning objective",
                                    "items": {
                                        "type": "string",
                                        "enum": all_verbs,
                                    },
                                },
                                "dependencies": {
                                    "type": "array",
                                    "description": "All learning objectives/knowledge/topics/subjects that must be learned before this one",
                                    "items": {
                                        "type": "string",
                                    },
                                },
                            },
                            "required": ["description", "knowledge_types", "cognitive_processes", "verbs"],
                        },
                    },
                },
                "required": ["learning_objectives"],
            },
        }
    ]
    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo-0613",
        messages=messages,
        functions=functions,
        # function_call="auto",  # auto is default, but we'll be explicit
        function_call={"name": "write_knowledge"}
    )
    response_message = response["choices"][0]["message"]

    # Step 2: check if GPT wanted to call a function
    if response_message.get("function_call"):
        # Step 3: call the function
        function_name = response_message["function_call"]["name"]
        function_args = json.loads(
            response_message["function_call"]["arguments"]
        )
        print(function_name)
        print(function_args)
        return function_args
```

```python
def json_to_mermaid_dag(json_data):
    mermaid_dag = "graph TD;\n\n    %% Nodes for Learning Objectives\n"
    
    # Define the nodes based on learning_objectives
    for obj in json_data['learning_objectives']:
        mermaid_dag += f"    lo{obj['id']}(\"{obj['id']} [{obj['type']}]: {obj['desc']}\")\n"

    mermaid_dag += "\n    %% Edges based on dependencies\n"
    
    # Define the edges based on dependencies
    for dep in json_data['dependencies']:
        node_id = dep['id']
        for dependency in dep['depends_on']:
            mermaid_dag += f"    lo{dependency} --> lo{node_id}\n"

    return mermaid_dag
```
