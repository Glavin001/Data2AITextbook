```python
def generate_questions_dataset(data):
    """
    Generate a dataset of out-of-context questions from the given unstructured text/passage.
    """
    final = []

    # print(data)
    # for row in data:
    for ri, row in enumerate(data):
        # print(row)
        id = row['id']
        source = row['source']
        context = row['story']
        questions = row['questions']
        answers = row['answers']

        # print(id)
        # print(source)
        # print(context)

        # add id, source, context to output
        # common = {}
        # common['id'] = id
        # common['source'] = source
        # common['context'] = context

        instruction = f'{context} {QUESTIONS_ACTION_TOKEN}'
        output = ''
        for qi, (q, a) in enumerate(zip(questions, answers)):

            paraphrases = q['paraphrase']

            # output = []
            output += f'Q{qi + 1}:\n'
            for pi, p in enumerate(paraphrases):
                output += f'{pi + 1}. {p}\n'

        qa = {
            'id': id,
            'source': source,
            'instruction': instruction,
            'output': output
        }

        final.append(qa)

    # return final
    final_dict = {
        'id': [qa['id'] for qa in final],
        'source': [qa['source'] for qa in final],
        'instruction': [qa['instruction'] for qa in final],
        'output': [qa['output'] for qa in final]
    }
    ds = Dataset.from_dict(final_dict)
    return ds

# generate_questions_dataset(ds).to_pandas()
# generate_questions_dataset(ds)
# limit dataset to 10 rows
# print(generate_questions_dataset([ds[1]])[0]['output'])

# save to .jsonl
generate_questions_dataset(ds).to_json('coqar-questions.jsonl', orient='records', lines=True)
```
